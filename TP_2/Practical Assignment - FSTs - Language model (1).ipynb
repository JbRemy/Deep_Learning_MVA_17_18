```python
# QUESTION a)

UPPER_lower_input_table = fst.SymbolTable(name="UPPER_lower_input")
for idx, word in enumerate(["<eps>", "A", "B"]):
    UPPER_lower_input_table.add_symbol(symbol=word, key=idx)

UPPER_lower_output_table = fst.SymbolTable(name="UPPER_lower_output")
for idx, word in enumerate(["<eps>", "b", "a"]):
    UPPER_lower_output_table.add_symbol(symbol=word, key=idx)
    
UPPER_lower = fst.Compiler(fst_type='vector')
print >> UPPER_lower, "0 1 1 2"
print >> UPPER_lower, "0 1 2 1"
print >> UPPER_lower, "1"
UPPER_lower_data = UPPER_lower.compile()
UPPER_lower_data.set_input_symbols(UPPER_lower_input_table)
UPPER_lower_data.set_output_symbols(UPPER_lower_output_table)
UPPER_lower_data.copy().closure()
```

#### Exercice 2) Edit Distance and Weighted Finite State Transducers (WFSTs)

It is possible to compute the usual [Levenshtein distance or Edit Distance](https://en.wikipedia.org/wiki/Levenshtein_distance) with Weighted Finite State Transducers. 
In the cell below, it is an example of the operations to compute the Edit Distance with only additions taken into account for basic sequences from this Vocabulary $\Sigma = \{0,1\}$.
```python
# QUESTION a)

alphabet = ['&#949;', 'h', 'o', 'm', 'l', 'a', 'c', 'g']
alphabet_table = fst.SymbolTable(name='Boolean')
for idx,element in enumerate(alphabet):
    alphabet_table.add_symbol(symbol=element,key=idx)

# Edit distance with FST: additions, substitutions and deletions.
add_fst = fst.Fst()
add_fst.reserve_states(1) 
s = add_fst.add_state()
add_fst.set_start(s)
for word_1 in range(len(alphabet)):
    for word_2 in range(len(alphabet)):
        add_fst.add_arc(s, fst.Arc(word_1, word_2, 1.0, s))
        
add_fst.set_final(s, 1)
add_fst.set_input_symbols(alphabet_table)
add_fst.set_output_symbols(alphabet_table);
```

```python
sequence1 = fst.Fst()
sequence1.reserve_states(3)
s = sequence1.add_state()
sequence1.set_start(s)
n = sequence1.add_state()
sequence1.add_arc(s, fst.Arc(4, 4, 1.0, n))
s = n
n = sequence1.add_state()
sequence1.add_arc(s, fst.Arc(5, 5, 1.0, n))
s = n
n = sequence1.add_state()
sequence1.add_arc(s, fst.Arc(6, 6, 1.0, n))
sequence1.set_final(n, 1)
sequence1.set_input_symbols(alphabet_table)
sequence1.set_output_symbols(alphabet_table)
```

```python
sequence2 = fst.Fst()
sequence2.reserve_states(5)
s = sequence2.add_state()
sequence2.set_start(s)
n = sequence2.add_state()
sequence2.add_arc(s, fst.Arc(4, 4, 1.0, n))
s = n
n = sequence2.add_state()
sequence2.add_arc(s, fst.Arc(5, 5, 1.0, n))
s = n
n = sequence2.add_state()
sequence2.add_arc(s, fst.Arc(3, 3, 1.0, n))
s = n
n = sequence2.add_state()
sequence2.add_arc(s, fst.Arc(5, 5, 1.0, n))
s = n
sequence2.set_final(n, 1)
sequence2.set_input_symbols(alphabet_table)
sequence2.set_output_symbols(alphabet_table)
```

```python
full_model = fst.compose(fst.compose(sequence1,add_fst),sequence2)
full_model
```

```python
fst.shortestpath(full_model)
```

```python
# QUESTION b)

#dict = {'coml','almanach'}, the text with mispells is T = 'cogml coml almanach lomh comll lamanach cocoml'

alphabet = np.unique(['&#949;'] + [_ for _ in 'coml'] + [_ for _ in 'almanach'] + ['g'])
dict_alphabet = {}
for _, letter in enumerate(alphabet):
    dict_alphabet[letter] = _
    
alphabet_table = fst.SymbolTable(name='Boolean')
for idx,element in enumerate(alphabet):
    alphabet_table.add_symbol(symbol=element,key=idx)

# Edit distance with FST: additions, substitutions and deletions.
dict_fst = fst.Fst()
dict_fst.reserve_states(1) 
s = dict_fst.add_state()
dict_fst.set_start(s)
count = 1
for word in ['coml', 'almanach']:
    c = s
    for _ in word:
        
        n = dict_fst.add_state()
        dict_fst.add_arc(c, fst.Arc(dict_alphabet[_], dict_alphabet[_], 1.0, n))
        c = n
        count += 1
        
    
    dict_fst.set_final(n, 1)

dict_fst.arcsort()
dict_fst.set_input_symbols(alphabet_table)
dict_fst.set_output_symbols(alphabet_table)
dict_fst
```

```python
sentence_fst = fst.Fst() 
s = sentence_fst.add_state()
sentence_fst.set_start(s)
count = 1
for word in 'cogml coml almanach lomh comll lamanach cocoml'.split():
    c = s
    for _ in word:
        n = sentence_fst.add_state()
        sentence_fst.add_arc(c, fst.Arc(dict_alphabet[_], dict_alphabet[_], 1.0, n))
        c = n
        count += 1
        
    
    sentence_fst.set_final(n, 1)

sentence_fst.arcsort()
sentence_fst.set_input_symbols(alphabet_table)
sentence_fst.set_output_symbols(alphabet_table)
```
